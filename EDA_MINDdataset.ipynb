# Importing the required modules and libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Loading the behaviors.tsv file into behaviours_df dataframe

```python
# The behaviors.tsv file contains the impression logs and users' news click histories. 
# It has 5 columns divided by the tab symbol:
# - Impression ID. The ID of an impression.
# - User ID. The anonymous ID of a user.
# - Time. The impression time with format "MM/DD/YYYY HH:MM:SS AM/PM".
# - History. The news click history (ID list of clicked news) of this user before this impression.
# - Impressions. List of news displayed in this impression and user's click behaviors on them (1 for click and 0 for non-click).
behaviors_path = os.path.join('./behaviors.tsv')
behaviours_df=pd.read_table(
    behaviors_path,
    header=None,
    names=['impression_id', 'user_id', 'time', 'history', 'impressions'])
```

# Loading the news.tsv file into a news_df dataframe 

```python
# The news.tsv file contains the detailed information of news articles involved in the behaviors.tsv file.
# It has 7 columns, which are divided by the tab symbol:
# - News ID
# - Category
# - Subcategory
# - Title
# - Abstract
# - URL
# - Title Entities (entities contained in the title of this news)
# - Abstract Entities (entities contained in the abstract of this news)
news_path = os.path.join('./news.tsv')
news_df=pd.read_table(news_path,
              header=None,
              names=[
                  'id', 'category', 'subcategory', 'title', 'abstract', 'url',
                  'title_entities', 'abstract_entities'
              ])
```

# Visualizing the distribution of impressions among users

```python
impressions_per_user = behaviours_df.groupby('user_id')['impression_id'].count()

# Plot the distribution of impressions per user
plt.figure(figsize=(10, 6))
sns.histplot(impressions_per_user, kde=True, bins=30)
plt.title('Distribution of Impressions per User')
plt.xlabel('Number of Impressions')
plt.ylabel('Frequency')
plt.show()

# Get the average number of impressions per user
avg_impressions = impressions_per_user.mean()
print(f'Average number of impressions per user: {avg_impressions:.2f}')
```

# Click Through Rate or 'CTR' from impressions
## if the impressions end with 1, they resulted in clicks 
### CTR is the ratio of how many of the total impressions actually resulted in clicks

```python
# Define a function to calculate CTR from the impressions
def calculate_ctr(impressions):
    impressions_list = impressions.split(' ')
    total_impressions = len(impressions_list)
    clicks = sum([1 for imp in impressions_list if imp.endswith('1')])
    return clicks / total_impressions if total_impressions > 0 else 0

# Apply the CTR calculation
behaviours_df['CTR'] = behaviours_df['impressions'].apply(calculate_ctr)

# Plot CTR distribution
plt.figure(figsize=(10, 6))
sns.histplot(behaviours_df['CTR'], kde=True, bins=30)
plt.title('Click Through Rate Distribution')
plt.xlabel('CTR')
plt.ylabel('Frequency')
plt.show()

# Calculate average CTR
avg_ctr = behaviours_df['CTR'].mean()
print(f'Average Click-Through Rate: {avg_ctr:.2%}')
```

# Visualize the distrubution of the avergage CTR per news category

```python
# Split impressions into news IDs and click values
behaviours_df['impression_List'] = behaviours_df['impressions'].apply(lambda x: x.split(' '))

# Create a DataFrame where each impression is a separate row
exploded_df = behaviours_df.explode('Impression_List')

# Split news ID and click status from Impression_List column
exploded_df[['News_ID', 'Click']] = exploded_df['Impression_List'].str.split('-', expand=True)
exploded_df['Click'] = exploded_df['Click'].astype(int)

# Merge with the news DataFrame to get categories
merged_df = pd.merge(exploded_df, news_df[['id', 'category']], left_on='News_ID', right_on='id', how='left')

# Calculate the average CTR by category
category_ctr = merged_df.groupby('category')['Click'].mean()

# Plot CTR by category
plt.figure(figsize=(20, 10))
sns.barplot(x=category_ctr.index, y=category_ctr.values, hue=category_ctr.index,palette='pastel')
plt.title('Click Through Rate (CTR) by News Category')
plt.xlabel('Category')
plt.ylabel('Average CTR')
plt.show()
```
